# APLICAÇÃO DO ALGORTIMO DE REDES NEURIAS ARTIFICIAIS.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import initializers
```

Leitura do dataset final.

```python
dataset = pd.read_csv('DATASET_FINAL_PARA_APLICAÇÃO_DOS_ALGORITIMOS_RNA.csv')
dataset.head()
```

```python
dataset.info()
```

Só mostrando que o alvo a ser predito está balanceado.

```python
dataset['target'].value_counts()
```

Aqui normalizamos o dataset, trazendo para uma escala de 0 a 1. Bem importante aqui em RNA's.

```python
normalizador = MinMaxScaler(feature_range=(0,1))

normalizador_features = normalizador.fit_transform(dataset.drop('target',axis=1))

dataset_normalizador = pd.DataFrame(normalizador_features,columns=dataset.columns[1:])

X_train, X_test, y_train, y_test = train_test_split(normalizador_features, dataset['target'], test_size=0.25)
```

### Criando a rede Neural e fazendo a modelagem

```python
%%time
    classificador = Sequential()
    
    classificador.add(Dense(units = 300, activation = 'relu', kernel_initializer = 'random_uniform', input_dim = 39))
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 300, activation = 'relu', kernel_initializer = 'random_uniform'))
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units = 300, activation = 'relu', kernel_initializer = 'random_uniform'))
    classificador.add(Dropout(0.2))
    
    classificador.add(Dense(units=1, activation='sigmoid'))
    
    otimizador = keras.optimizers.Adam(lr = 0.0001, decay = 0.001, clipvalue = 0.3)
    classificador.compile(optimizer = otimizador, loss = 'binary_crossentropy', metrics = ['binary_accuracy']) 
    
    classificador.fit(X_train, y_train, epochs = 1000, batch_size = 50)
```

Aqui, extraimos as predições feitas nos dados de teste.

```python
predictions = classificador.predict(X_test)
predictions = (predictions > 0.5)
```

## CLASSIFICAÇÃO. 

Classification report nos traz as principais métricas de desempenho do algoritmo nesses dados.

```python
print(classification_report(y_test, predictions))
```

## Importar modelo para usarmos depois no processo de inferência.

```python
classificador_json = classificador.to_json() 

with open('classificador_MLP.json', 'w') as json_file:
    json_file.write(classificador_json) 
    
classificador.save_weights('classificador_MLP.h5') 
```
