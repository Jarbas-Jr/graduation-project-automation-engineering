# ESSE DATASET VISA APENAS CONVERTER PARA VALOR NÚMERIO OS LABELS DO TARGET, JÁ QUE A METODOLOGIA DAS RNA'S NAO ACEITA STRINGS.

## GOOD --> 1
## BAD --> 0

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('DATASET_FINAL_PARA_APLICAÇÃO_DOS_ALGORITIMOS.csv')
dataset.head()
```

```python
dataset.target[dataset.target == 'Bad'] = 0
dataset.target[dataset.target == 'Good'] = 1
```

```python
dataset.head()
```

```python
dataset['target'] = dataset['target'].astype(float)
```

```python
dataset.info()
```

```python
dataset.to_csv('DATASET_FINAL_PARA_APLICAÇÃO_DOS_ALGORITIMOS_RNA.csv', index = False)
```
